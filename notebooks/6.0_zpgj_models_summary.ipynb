 En este proyecto nos enfocamos en la clasificación de emociones utilizando diferentes técnicas de aprendizaje automático, como Support Vector Machines (SVM), Bosque aleatorio (RF) y K-Nearest Neighbors (K-NN). Usando este dataset, se logró clasificar canciones de emociones como alegría, tristeza, enojo y relajación. ### **2. La importancia de la reproducibilidad en el aprendizaje automático**

La reproducibilidad en el aprendizaje automático es crucial porque garantiza que los resultados de los experimentos o modelos puedan ser replicados por otros usuarios o en diferentes entornos. Esto fomenta la transparencia, la fiabilidad y la confianza en los resultados, permitiendo validar los modelos y asegurarse de que no están sobreajustados a conjuntos de datos específicos.

### **3. MLflow para el seguimiento y la reproducibilidad de los experimentos.**

MLflow permitió registrar automáticamente los parámetros, métricas, artefactos y versiones de código de cada experimento de ML. Al almacenar esta información en una base de datos centralizada, facilita la comparación de diferentes ejecuciones, lo que permite analizar su rendimiento y resultados en detalle. También proporcionó un registro centralizado para administrar versiones de modelos, con capacidades para realizar un seguimiento de las versiones aprobadas. Esto garantiza que se pueda volver a una versión específica del modelo cuando sea necesario, mejorando la trazabilidad y la reproducibilidad en el ciclo de vida del modelo.


### **4. Reflexión sobre los desafíos enfrentados y cómo se superaron**

A medida que los experimentos se vuelven más numerosos, mantener el control de versiones de los modelos se vuelve más compleja y por lo tanto es crucial establecer un flujo de trabajo en MLFlow claro para asociar correctamente cada conjunto de datos con sus respectivos modelos y resultados.
